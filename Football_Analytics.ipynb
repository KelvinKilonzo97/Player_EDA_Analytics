```python
def sample_data(df):
    A = df.National_Position.replace({'':"Other","Sub":"Other"}).fillna("Other")
    A1 = df.Work_Rate.str.split("/" , expand = True).iloc[:,1].rename('Work_rate1')
    A2 = df.Work_Rate.str.split("/" , expand = True).iloc[:,0].rename('Work_rate2')
    A3 = df.Club.replace({'Free agent':'Free Agents'})
    A4 = df.Height.str.slice(0,3).astype(int).rename("Player_Height")
    A5 = df.Height.str.slice(0,3).astype(int).rename("Player_Height")
    def limit_country(val):
      if val in {"England","Argentina","Spain","France","Brazil","Kenya"}:
        return val
      return "Other"
    A6 = df.Nationality.apply(limit_country)
    A7 = df.Rating
    return pd.concat([A6,A,A4,A5,A1,A2,A3,A7],axis = 1)
Sample = sample_data(df)
```
